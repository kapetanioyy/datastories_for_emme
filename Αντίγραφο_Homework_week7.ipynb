3.7. Συμπεράσματα από την πρώτη επισκόπηση του dataframe.

Έχουμε ένα dataframe που περιλαμβάνει δεδομένα για 979 ταινίες από το IMDB.

Στα δεδομένα περιλαμβάνονται:
- ο τίτλος της κάθε ταινίας (στήλη 'title')
- η βαθμολογία της (στήλη 'star_rating')
- o χαρακτηρισμός του περιεχομένου της (στήλη 'content_rating')
- το είδος της ταινίας (στήλη 'genre')
- η διάρκειά της (στήλη 'duration')

Οι στήλες 'star_rating' και 'duration' περιέχουν αριθμητικά δεδομένα.

Όμως, οι στήλες 'content_rating' και 'genre' περιέχουν συγκεκριμένες τιμές (πολλές ταινίες ανήκουν στο ίδιο είδος και έχουν την ίδια σήμανση).

4.5. Ποιες είναι οι 5 ταινίες δράσης με τη μεγαλύτερη βαθμολογία;

```python
# θέλω το dataframe να εμφανίζει μόνο τις γραμμές που είναι True στην παραπάνω συνθήκη (δηλαδή να φιλτραριστεί το dataframe)
# γράφω το ονομα του dataframe και μέσα σε αγκύλες βάζω το φίλτρο
movies_df[movies_df['genre'] == 'Action']
```

Τώρα μπορώ να τις ταξινομήσω ως προς το star_rating σε φθήνουσα σειρά και να πάρω τις 5 πρώτες.

```python
movies_df[movies_df['genre'] == 'Action'].sort_values(by="star_rating", ascending=False).head()[["genre", "title", "star_rating"]]
```
